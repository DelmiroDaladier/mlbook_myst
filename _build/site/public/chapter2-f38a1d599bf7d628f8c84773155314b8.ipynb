```python
import altair as alt
import pandas as pd

ranking = [1, 1, 1, 0, 1, 0, 0, 1, 0, 0, 0]
x = [0, 0, 0, 300, 300, 300, 600, 600, 600, 900, 1200]
y = [0, 200, 400, 400, 600, 800, 800, 1000, 1200, 1200, 1200]

df = pd.DataFrame({'x': x, 'y': y, 'ranking': ranking})

grid_step = 120
h = 1200
w = 1200

vgrid = pd.DataFrame({'x': list(range(grid_step, w + 1, grid_step))})
hgrid = pd.DataFrame({'y': list(range(grid_step, h + 1, grid_step))})

vlines = alt.Chart(vgrid).mark_rule(strokeDash=[2, 2], color='gray').encode(x='x')
hlines = alt.Chart(hgrid).mark_rule(strokeDash=[2, 2], color='gray').encode(y='y')


main_line = alt.Chart(df).mark_line(strokeDash=[5, 5]).encode(x='x', y='y')
points = alt.Chart(df).mark_point().encode(x='x', y='y')

segments = pd.DataFrame({
    'x': [x[0], x[2], x[2], x[8], x[8], x[10]],
    'y': [y[0], y[2], y[2], y[8], y[8], y[10]],
    'group': [1, 1, 2, 2, 3, 3],
    'width': [4, 4, 2, 2, 4, 4]
})

red_lines = alt.Chart(segments).mark_line().encode(
    x='x',
    y='y',
    detail='group:N',
    strokeWidth='width:Q',
    color=alt.value('red')
)

x_axis = alt.Axis(
    values=list(range(0, 1201, 120)),
    labelExpr='{"360": "0.25", "600": "0.50", "960": "0.75", "1200": "1.00"}[datum.value] || ""',
    title='False positive rate'
)

y_axis = alt.Axis(
    values=list(range(0, 1201, 120)),
    labelExpr='{"120": "0.17", "240": "0.33", "480": "0.50", "600": "0.67", "840": "0.83", "960": "1.00"}[datum.value] || ""',
    title='True positive rate'
)

chart = (vlines + hlines + main_line + points + red_lines).properties(
    width=800,
    height=600
).configure_axis(
    grid=False
).encode(
    x=alt.X('x:Q', axis=x_axis),
    y=alt.Y('y:Q', axis=y_axis)
)

chart.display()
```
